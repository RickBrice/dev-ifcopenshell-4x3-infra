# 002 Plotting

Develop plots for visually checking vertical and cant alignment calculation

```python
import pathlib
import os

import ifcopenshell
import ifcopenshell.geom
import ifcopenshell.ifcgeom


index = 3
geometry = True

test_name = f"UT_AWC_{index}"
awc_path = os.path.join(
    pathlib.Path.home(),
    "src",
    "IFC-Rail-Sample-Files",
    "1_Alignment with Cant (AWC)",
    test_name,
    "IFC reference files",
)
if index < 6:
    awc_path = os.path.join(awc_path, "RC4")

if index == 7:
    test_file = f"{test_name}_GeometryGym.ifc"
else:
    if geometry:
        test_file = f"{test_name}.ifc"
    else:
        test_file = f"{test_name}_no_geometry.ifc"
in_file = os.path.join(awc_path, test_file)

model = ifcopenshell.open(in_file)

ent = model.by_type("IfcAlignment")[0]

align = ifcopenshell.ifcgeom.Alignment().from_entity(ent)
align.plot(title=test_file, interactive=True)


```
